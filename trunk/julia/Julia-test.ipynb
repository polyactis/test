# Plot runtime vs the number of data points. eGADA vs GADA.

```julia
using StatsPlots
```

```julia
x = [320_000, 1_280_000, 6_400_000, 12_800_000]
y_GADA = [2.85, 13.88, 110.74, 326.35]
y_eGADA = [1.88, 8.52,  53.6, 102.63]
plot(x, y_eGADA; xlabel="#data points in input", 
    ylabel="Runtime (in seconds)", label = "eGADA", markershape=:circle, 
    left_margin=5*Plots.mm, bottom_margin=5*Plots.mm, dpi=500, size=(900,500) )

plot!(x, y_GADA; xlabel="#data points in input", 
    ylabel="Runtime (in seconds)", label = "GADA", markershape=:circle, style=:dash)
```
